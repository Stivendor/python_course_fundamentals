```python
import pandas as pd

df = pd.read_csv("../data/gemini/electronic_sales.csv", index_col="Transaction_ID")
df
```

```python
# Quick Inspection: Display the first 5 rows of the DataFrame and its general information (data types, non-null values).
df.head()
```

```python
# Type Conversion: Convert the Sale_Date column to the datetime data type.
df["Sale_Date"] = pd.to_datetime(df["Sale_Date"])
df
```

```python
# Null Check: Check if there are any null values in the DataFrame. (Although this CSV is clean, this is a crucial step).
df.isnull().sum()
```

```python
# Descriptive Statistics: Show the descriptive statistics for the numerical columns.
df.describe()
```

```python
# Total Calculation: Create a new column named Total_Sale which is the result of multiplying Quantity by Unit_Price.
df["Total_Sale"] = df["Quantity"] * df["Unit_Price"]
df
```

```python
# Day Extraction: Create a column named Day_of_Week that contains the day of the week (e.g., 'Monday', 'Tuesday', etc.) derived from the Sale_Date column.
df["Day_of_Week"] = df["Sale_Date"].dt.day_name()
df
```

```python
df
```

```python
# Conditional Filter: Select and display only the rows where the Quantity sold is greater than 1.
df[df["Quantity"] > 1] 
```

```python
# Compound Filter: Select transactions that belong to the 'Smartphones' Category and had a Payment_Method of 'Card'.
df[(df["Category"].str.contains("Smartphones")) & (df["Payment_Method"].str.contains("Card"))]
```

```python
# Specific Column Selection: Create a new DataFrame containing only the columns Product, Category, and Total_Sale.
new_df = df[["Product","Category", "Total_Sale"]]
new_df
```

```python
# Sales by City: Calculate the sum of the Total_Sale for each City. Which city has the highest sales?
sales_by_city = df.groupby("City").sum(numeric_only=True)
sales_by_city.sort_values(by="Total_Sale", ascending=False)
```

```python
# Count by Category: Count how many transactions there are for each Category.
count_by_category = df.groupby("Category").sum(numeric_only=True)
count_by_category.sort_values(by="Quantity", ascending=False)
```

```python
# Multiple Aggregation: Group by Category and calculate the minimum, maximum, and average of the Unit_Price.
count_by_category["Unit_Price"].describe()
# max = 3199
# min = 199
# mean/avg = 1207
```

```python
# Most Expensive Product: Sort the DataFrame by Unit_Price in descending order and display the most expensive product sold.
mep = df["Unit_Price"].sort_values( ascending = False)
mep.head()
```
