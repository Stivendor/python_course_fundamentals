```python
import pandas as pd
df = pd.read_csv("../data/dataset.csv", index_col="id") # index_col is used to give a new index, in this case "id"
df
```

```python
# head function is used to show the first five rows
df.head()
```

```python
# head function is used to show the last five rows
df.tail(7) # we can specify the number of elements
```

```python
# we can use the function .describe to show all the descriptive stadistic of the cvs file
df.describe()
```

```python
# we always want to clean NaN values, for this we can use the function dropna
df_clean = df.dropna()
df_clean.head()
```

```python
# but if we don't want to clean these values, we can fill them with a specific value using fillna
df_filled = df.fillna(0)
df_filled.head()
```

```python
# also, we can specify the value on each camp with a dictionary
df_filled_dict = df.fillna({"retweets":0,"mentions":-1})
df_filled_dict.head()
```

```python
# get a column
df["favorites"]
```

```python
# get more than one column, we put a list 
df[["favorites","user"]]
```

```python
# to get a row by index, we use the function iloc
df.iloc[3] # get the row with index 3
```

```python
# to get a range of rows by index, we use the function iloc
df.iloc[0:3] # get the info of the row 0, 1, 2
```

```python
# and also we can specify the index in a list 
df.iloc[[1,2,20]]
```

```python
# if we want to filter by id, we should use loc (the same but without i)
df.loc[[183722, 183741]]
```

```python
# to filter both (rows and columns) we do this
df.loc[[183722, 183741], ["favorites","country"]]
```

```python
# conditional filter for number columns
df[(df["favorites"] > 400) & (df["mentions"] > 20)]
```

```python
# filter for string columns
df[df["full_text"].str.contains("original")]
```

```python
# data transformation, for example, we can create the next function
import random
def calculate_earnings(retweets):
    earning = retweets * random.randint(3, 5)
    return earning
df["earning"] = df["retweets"].apply(calculate_earnings)
df.head(10)

# if we want to calculate the popularity of each user based on the number of followers and followees, we can do this
def popularity(row):
    result = row["followees"]/row["followers"] if row["followers"] > 0 else 0
    return result
df["popularity"] = df.apply(popularity, axis=1) # axis=1 means that we are applying the function to each row
df.head()

```

```python
# data agrupation by a specific column, for example, country
df_grouped = df.groupby("country").mean(numeric_only=True) # we can use other functions like sum
df_grouped
```

```python
# agregation with multiple functions
df_agg = df.groupby("country").agg({
    "favorites": ["mean", "sum"],
    "retweets": ["mean", "sum"]
})
df_agg
```

```python
# to graphic the data
import matplotlib.pyplot as plt

df_agg["favorites"].plot(kind="bar")
plt.show()
```

```python
df.plot(kind="scatter", x="favorites", y="retweets")
plt.show()
```

```python
# saving a modify data frame
df_agg.to_csv("../data/output.csv")
```
